```python
#We can see that we don’t want to print out all the tokens, but we can use the Python list
#“slice” notation to print the first 30 tokens in the list of sorted tokens:

sorted(set(text3))[:30]
```

```python
#The ratio of the total number of unique tokens and the total number of tokens indicates the
#lexical richness of the text. 
len(set(text3)) / len(text3)
```

```python
#Now let’s search for and count occurrences of particular words and compare that to the
#total number of words.

text3.count("smote")
```

```python
#Compute the fraction of the number of occurrences of the word compared with the total
#number of words and then multiply by 100 to get a percentage.

100 * text3.count('smote')/len(text3)
```

```python
#How does this compare with a more common word, such as the word “and”?

100 * text3.count('and')/len(text3)
```
